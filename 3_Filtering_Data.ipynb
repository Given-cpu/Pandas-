```python
import pandas as pd
bio = pd.read_csv('bios.csv')
bio.info()
```

```python
bio.head()
```

```python
bio.loc[bio['height_cm'] > 220, ['name']]
```

```python
bio[bio['height_cm'] > 220]['name']
```

```python
bio[(bio['height_cm'] > 220) & (bio['born_country'] == 'CHN')]
```

```python
bio[bio['born_date'].str.contains('08-24', na=False)]
```

```python
bio[bio['name'].str.contains('zane', case=False)]
```

```python
bio[bio['name'].str.contains(r'son$|sen$',case=False,na=False)]
```

```python
bio[bio['born_country'].isin(['CHN','GBR','USA']) & bio['name'].str.startswith('Kate')]
```

```python
bio.query('height_cm > 220 and born_country == "CHN"')
```
